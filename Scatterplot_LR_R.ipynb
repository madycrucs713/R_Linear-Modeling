```r
library(ggplot2)
data = read.csv('regrex1.csv')
ggplot(data, aes(x = x, y = y)) +
  geom_point(color = "blue") +
  geom_smooth(method = "lm", se = FALSE, color = "red") +
  labs(title = "Scatter Plot with Linear Model using R",
       x = "X axis",
       y = "Y axis")
model = lm(data)
print(model)
```
